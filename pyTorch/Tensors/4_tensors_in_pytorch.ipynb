Inplace Operations

```python
m = torch.rand(2,3)
n = torch.rand(2,3)

print(m)
print(n)
```

```python
# adding m and n ; storing result in m
m.add_(n)
```

**Copying a tensor**

```python
a = torch.rand(2,3)
a
```

```python
b = a
b
```

```python
a[0][0] = 0
print(a)
print(b)
```

```python
print(id(a))
print(id(b))
```

```python
b=a.clone()
b
```

```python
# now both tensors are refering to the different memory
print(id(a))
print(id(b))
```
