**Tensor Shapes**

```python
x = torch.tensor([[1,2,3],[4,5,6]])
print(x)
```

```python
x.shape
```

```python
torch.empty_like(x)
```

```python
torch.zeros_like(x)
```

```python
torch.ones_like(x)
```

```python
torch.rand_like(x)
# reason:rand generates the floats so type cast
```
