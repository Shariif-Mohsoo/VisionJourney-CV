**Scalar Operations**

```python
x = torch.rand(2,2)
print(x)
```

```python
# addition
print(x+2)
# subtration
print(x-2)
# multiplication
print(x*2)
# division
print(x/3)
# int division
print((x * 100) // 3)
# mod
print((x * 100) % 3)
# power
print(x**2)
```

**Element Wise Operation**

```python
a = torch.rand(2,3)
b = torch.rand(2,3)

print(a)
print(b)
```

```python
# add
print(a+b)
# sub
print(a-b)
# multiplication
print(a*b)
# division
print(a/b)
# mod
print(a%b)
```

```python
c = torch.tensor([1,-2,3,-4])
print(c)
```

```python
# abs
torch.abs(c)
```

```python
# negative
torch.neg(c)
```

```python
d = torch.tensor([1.9,2.3,3.7,4.4])
print(d)
```

```python
torch.round(d)
```

```python
# ceil
torch.ceil(d)
```

```python
# floor
torch.floor(d)
```

```python
# clmap defines the range for entire elements in tensor (b/w min and max)
torch.clamp(d,min=2,max=3)
```

**Reduction operation** Mean reducing the entire tensor into single number like array reduce method in js.

```python
e = torch.randint(size=(2,3),low=0,high=10,dtype=torch.float32)
e
```

```python
# sum
print(torch.sum(e))
# sum along columns
print(torch.sum(e,dim=0))
# sum along rows
print(torch.sum(e,dim=1))
```

```python
# mean
print(torch.mean(e))
# mean along col
torch.mean(e,dim=0)
```

```python
# median
print(torch.median(e))
# max and min
print(torch.max(e))
print(torch.min(e))
# product
print(torch.prod(e))
```

```python
# standard deviation
torch.std(e)
```

```python
# variance
torch.var(e)
```

```python
# argmax => return the position of the max element
torch.argmax(e)
```

```python
# argmin => return the position of the min element
torch.argmin(e)
```

**Matrix Operations**

```python
f = torch.randint(size=(2,3),low=0,high=10)
g = torch.randint(size=(3,2),low=0,high=10)

print(f)
print(g)
```

```python
# matrix multiplication
torch.matmul(f,g)
```

```python
# dot product
v1 = torch.tensor([1,2,3])
v2 = torch.tensor([3,2,1])
torch.dot(v1,v2)
```

```python
# transpose
torch.transpose(f,0,1)
```

**Comparison Operations**

```python
i = torch.randint(size=(2,3),low=0,high=10)
j = torch.randint(size=(2,3),low=0,high=10)

print(i)
print(j)
```

```python
# greater than
print(i>j)
# less than
print(i<j)
# equal to
print(i==j)
# not equal to
print(i!=j)
# greater than equal to
print(i>=j)
# less than equal to
print(i<=j)
```

**Special functions**

```python
k = torch.randint(size=(2,3),low=0,high=10)
k
```

```python
# log
print(torch.log(k))
# exp
print(torch.exp(k))
# sqrt
print(torch.sqrt(k))
# sigmoid
print(torch.sigmoid(k))
# softmax
print(torch.softmax(k.to(torch.float32),dim=0))
```

```python
# relu -> converts the negative value to zero
torch.relu(k)
```
