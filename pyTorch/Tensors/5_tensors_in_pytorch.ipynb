```python
import torch
print(torch.__version__)
```

```python
if torch.cuda.is_available():
  print("GPU is available")
  print(f"Using GPU: {torch.cuda.get_device_name(0)}")
else:
  print("GPU not available. Using CPU")
```

Tensor Operations on GPU

```python
device = torch.device('cuda')
print(device)
```

```python
# creating a new tensor on GPU
torch.rand((2,3),device=device)
```

```python
a = torch.rand(2,3)
a
```

```python
# moving an existing tensor to GPU
b = a.to(device)
b
```

```python
# now operations will run on gpu
b+5
```

**Reshaping Tensors**

```python
a = torch.ones(4,4)
a
```

```python
# reshape
a.reshape(2,2,2,2)
```

```python
# flatten
a.flatten()
```

```python
b = torch.rand(2,3,4)
b
```

```python
# permute => adjusting the dimension as per the need
b.permute(2,0,1)
```

```python
# typical image size
c = torch.rand(226,226,3)
# unsqueeze => adds the new dimension in the given location
c.unsqueeze(0).shape
```

```python
# squeeze => removes the dimension from the given location
d = torch.rand(1,20)
print(d)
d.squeeze(0)
```
