```python
import torch
import torch.nn.functional as F
```

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
```

```python
w = torch.tensor(1.0,requires_grad=True)
b = torch.tensor(0.0,requires_grad=True)
```

```python
z = w*x + b
```

```python
y_pred = torch.sigmoid(z)
```

```python
loss = F.binary_cross_entropy(y_pred, y)
```

```python
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```

**Vector Input Tensor**

```python
x = torch.tensor([1.0,2.0,3.0],requires_grad=True)
```

```python
y = (x**2).mean()
```

```python
print("Input ",x)
print("Output ",y)
```

```python
y.backward()
print(x.grad)
```

```python
# clearing grad
x = torch.tensor(2.0,requires_grad=True)
```

```python
y = x**2
```

```python
y.backward()
```

```python
x.grad
```

```python
x.grad.zero_()
```

**How to disable gradient tracking**

```python
x = torch.tensor(2.0,requires_grad=True)
```

```python
y = x ** 2
y
```

```python
y.backward()
```

```python
x.grad
```

```python
# option 1 - requires_grad_(False)
# option 2 - detach()
# option 3 - torch.no_grad()
```

```python
x.requires_grad_(False)
```

```python
y = x ** 2
```

```python
y
```

```python
y.backward()
```

```python
x = torch.tensor(2.0,requires_grad=True)
```

```python
x
```

```python
z = x.detach()
z # now no gradient function
```

```python
x = torch.tensor(2.0,requires_grad=True)
```

```python
with torch.no_grad():
  y = x ** 2
print(x)
print(y)
```
