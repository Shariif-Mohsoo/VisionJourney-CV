```python
import torch
```

```python
# find derivate of y = x^2 and x = 3
x = torch.tensor(3.0,requires_grad=True)
# requires_grade = true mean place track of all ops on tensor to calculate derivate
```

```python
y = x**2
print(x)
print(y)
```

```python
#  calculate derivate
y.backward()
```

```python
# printing the derivate
x.grad
```

```python
# Example: y = x^2 , z= sin(y) Find derivate with the help of chain rule.
x = torch.tensor(3.0,requires_grad=True)
```

```python
y = x ** 2
```

```python
z = torch.sin(y)
```

```python
print(x)
print(y)
print(z)
```

```python
# derivate calculation
z.backward()
```

```python
# printing the derivate result
x.grad
```
